
````python
# define the arrays
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])

# Using the function outer in numpy
np_outer = np.outer(u, v)
print('Using the function outer in numpy:')
print(np_outer)
print()

# Using a nested for loop or list comprehension
loop_outer = np.array([[u_i*v_j for v_j in v] for u_i in u])
print('Using a nested for loop or list comprehension:')
print(loop_outer)
print()

# Using numpy broadcasting operations
broadcasting_outer = u.reshape(4, 1) * v # another way u[:, np.newaxis] * v
print('Using numpy broadcasting operations:')
print(broadcasting_outer)
````

4\. Use np.linspace to create an array of 100 numbers between 0 and 2Ï€ (includsive).

  * Extract every 10th element using slice notation
  * Reverse the array using slice notation
  * Extract elements where the absolute difference between the sine and cosine functions evaluated at that element is less than 0.1

````python
# creat the array of angles
theta = np.linspace(start=0.0, stop=2*np.pi, num=100)
print(theta, '\n')

# extract every 10th element
theta_10th = theta[::10].copy()
print('every 10th element:\n', theta_10th, '\n')

# reversing
theta_rev = theta[::-1].copy()
print('reversed:\n', theta_rev, '\n')

# absolute difference between sin and cos less than 0.1
close_elements = theta[np.abs(np.sin(theta) - np.cos(theta)) < 0.1].copy()
print('close elements in sin and cos values:\n', close_elements, '\n')

# Make a plot showing the sin and cos functions and indicate where they are close

# plot sin and cos values
plt.plot(theta, np.sin(theta), label='Sin')
plt.plot(theta, np.cos(theta), label='Cos')

# plot close values
plt.scatter(close_elements, np.sin(close_elements), label='Close Points(Sin)')
plt.scatter(close_elements, np.cos(close_elements), label='Close Points(Cos)')

# make the plot more readable
plt.legend()
plt.title('Sin and Cos plot')
plt.xlabel('Theta')
plt.ylabel('F(Theta)')
plt.show()
````

6\. Use broadcasting to create a grid of distances

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles
The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Construct a 2D grid of distances among each city along Route 66
  * Convert that in km (those savages...)

````python
# name of the cities
cities = ['Chicago', 'Springfield',
          'Saint-Louis', 'Tulsa',
          'Oklahoma City', 'Amarillo',
          'Santa Fe', 'Albuquerque',
          'Flagstaff', 'Los Angeles']

# creat an array of positions
mile_positions = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])

# create a 2D grid of distances using broadcasting
mile_grid = np.abs(mile_positions - mile_positions[:, np.newaxis])

# Conversion factor from miles to kilometers
miles_to_km = 1.6 #1.60934

# Convert distances to kilometers
km_grid = mile_grid * miles_to_km

# Print the grid of distances in miles and kilometers
print("Grid of Distances (Miles):")
print(mile_grid)
print("\nGrid of Distances (Kilometers):")
print(km_grid)

# image size
plt.figure(figsize=(14, 5))

# Plot the mile_matrix as an image (subplot 1)
plt.subplot(1, 2, 1)
plt.imshow(mile_grid, cmap='viridis')
plt.colorbar().set_label('miles')
plt.xticks(np.arange(len(cities)), cities, rotation=70)
plt.yticks(np.arange(len(cities)), cities)
plt.title('Distances in Miles')

# Plot the km_matrix as an image (subplot 2)
plt.subplot(1, 2, 2)
plt.imshow(km_grid, cmap='cividis')
plt.colorbar().set_label('km')
plt.xticks(np.arange(len(cities)), cities, rotation=70)
plt.yticks(np.arange(len(cities)), cities)
plt.title('Distances in KM', ha='center')

plt.tight_layout()
plt.show()
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
def prime_number_sieve(N):
    
    # Constract a boolean array, the mask
    mask = np.ones(N+1, dtype=bool)

    # Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
    for i in range(2, N+1):
        mask[i*2: N+1: i] = False
        
    # Apply the mask to obtain an array of ordered prime numbers
    primes = np.arange(0, N+1)[mask][2:]
    
    return primes

# Test for N = 99
print('Prime numbers up to 99 using Prime numbers sieve:')
print(prime_number_sieve(99))
%timeit prime_number_sieve(99)
print()

# Check how the performance scales with N
time_taken_sieve = []
N_s = [100*i for i in range(1, 11)]
for N in N_s:
    time_taken_sieve.append(timeit.timeit(lambda: prime_number_sieve(N), number=10000))

# plot
# image size
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(N_s, time_taken_sieve, 'o')
plt.xlabel('N')
plt.ylabel('Time')
plt.title('prime_number_sieve(N)')

# Implement the optimization suggested in the sieve of Eratosthenes
def sieve_of_eratosthenes(N):
    
    # Constract a boolean array, the mask
    mask = np.ones(N+1, dtype=bool)

    # Set the multiples of each number starting from 2 in the mask
    for i in range(2, int(np.sqrt(N))+1):
        if mask[i]:
            mask[i*i : N+1 : i] = False

    # Extract the prime numbers using the mask
    primes = np.arange(0, N+1)[mask][2:]

    return primes

# Test for N = 99
print('Prime numbers up to 99 using sieve_of_eratosthenes')
print(sieve_of_eratosthenes(99))
%timeit sieve_of_eratosthenes(99)
print()

# Check how the performance scales with N
time_taken_sieve_era = []
N_s_era = [100*i for i in range(1, 11)]
for N in N_s_era:
    time_taken_sieve_era.append(timeit.timeit(lambda: sieve_of_eratosthenes(N), number=10000))


# plot
plt.subplot(1, 2, 2)
plt.plot(N_s_era, time_taken_sieve_era, 'o')
plt.xlabel('N')
plt.ylabel('Time')
plt.title('sieve_of_eratosthenes')

plt.tight_layout()
plt.show()
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
# np.random.choice([1, -1], size=(1000, 200), p=[0.5, 0.5])
````

````python
# Take 1000 walkers and let them walk for 200 steps
walkers = 1000
steps = 200

# Use randint to create a 2D array of size walkers x steps with values -1 or 1
walker_steps = np.random.randint(low=0, high=2, size=(walkers, steps), dtype=int) * 2 - 1

# Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
walker_distances = np.cumsum(walker_steps.copy(), axis=1)

# Take the square of that 2D array (elementwise)
walker_distances = np.square(walker_distances)

# Compute the mean of the squared distances at each step (i.e. the mean along the columns)
walker_distances = np.mean(walker_distances, axis=0)

# Plot the average distances (sqrt(distance**2)) as a function of time (step)
walker_distances = np.sqrt(walker_distances)
steps_range =  np.arange(1, steps+1)
plt.plot(steps_range, walker_distances, label='Simulation')
plt.plot(steps_range, np.sqrt(2*steps_range/np.pi), label='Theory')
plt.title('Average Distance from Origin over Time')
plt.xlabel('Time (Step)')
plt.ylabel('Average Distance')
plt.legend()
plt.grid()
plt.show()
````

````python
# Load the data (use an appropriate numpy method) into a 2D array
data = np.loadtxt('populations.txt').astype(int)
data
````

````python
# Create arrays out of the columns, the arrays being (in order): year, hares, lynxes, carrots
year = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

# Plot the 3 populations over the years
plt.figure(figsize=(10, 6))
plt.plot(year, hares, label='hares', marker='.')
plt.plot(year, lynxes, label='lynxes', marker='.')
plt.plot(year, carrots, label='carrots', marker='.')
plt.xticks(year, rotation=45)
plt.xlabel('year')
plt.ylabel('population')
plt.title('Populations of Hares, Lynxes, and Carrots Over the Years')
plt.legend()
plt.grid()
plt.show()
````

````python
# Compute the main statistical properties of the dataset (mean, std, correlations, etc.)
print("Statistical Properties:\n")

# mean
mean_hares = np.mean(hares)
mean_lynxes = np.mean(lynxes)
mean_carrots = np.mean(carrots)
print(f"Mean hares: {mean_hares}, Mean lynxes: {mean_lynxes}, Mean carrots: {mean_carrots}\n")

# std
std_hares = np.std(hares)
std_lynxes = np.std(lynxes)
std_carrots = np.std(carrots)
print(f"std hares: {std_hares}, std lynxes: {std_lynxes}, std carrots: {std_carrots}\n")

# correlations
correlation_hl = np.corrcoef(hares, lynxes)[0, 1]
correlation_lc = np.corrcoef(lynxes, carrots)[0, 1]
correlation_ch = np.corrcoef(carrots, hares)[0, 1]

print(f"Correlation Hares-Lynxes: {correlation_hl}")
print(f"Correlation Lynxes-Carrots: {correlation_lc}")
print(f"Correlation Carrots-Hares: {correlation_ch}")

# Calculate correlations between columns in data matrix
correlation_matrix = np.corrcoef(data[:, 1:], rowvar=False)

# Print the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)
````

````python
# Which species has the highest population each year?
max_population_species = np.argmax(data[:, 1:], axis=1)
species_names = ['Hares', 'Lynxes', 'Carrots']
max_population_species_names = np.array(species_names)[max_population_species]

# Print the species with the highest population each year
print("\nSpecies with the Highest Population Each Year:")
for yr, species in zip(year, max_population_species_names):
    print(f"Year {int(yr)}: {species}")
````
